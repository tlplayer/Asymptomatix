```python

import json

with open('input/fake_CT1000.txt') as f:
  data = json.load(f)

print(data[0].keys())
```

```python
unique_locations = {}

location_id = 0

for patient in data:
    for i in range(len(patient["lat"])):        
        unique_location_tuple = (patient["lat"][i], patient["lon"][i])
        if unique_location_tuple not in unique_locations:
            unique_locations[unique_location_tuple] = location_id
            location_id = location_id+1
```

```python
len(unique_locations)
```

```python
for key,value in unique_locations.items():
    print(str(value) + ","+ str(value) + "," + str(key[0]/10000) + "," + str( key[1]/10000)  )
#     print(str(key[1]/10000) )
```

```python
flow_counts = {}


for patient in data:
    for i in range(1, len(patient["lat"])): 
        
        k = i-1
        
        
        orig_location_tuple = (patient["lat"][k], patient["lon"][k])
        
        orig_location_tuple_id = unique_locations[orig_location_tuple]
        
        
        dest_location_tuple = (patient["lat"][i], patient["lon"][i])
        
        dest_location_tuple_id = unique_locations[dest_location_tuple]
        
        flow_tuple = (orig_location_tuple_id,dest_location_tuple_id )
        
        if flow_tuple not in flow_counts:
            
            flow_counts[flow_tuple] = 1
        else:
            flow_counts[flow_tuple] =  flow_counts[flow_tuple]+1

        
#        print(orig_location_tuple_id, dest_location_tuple_id, 1)
#         print(dest_location_tuple_id)
```

```python
from random import randint

for key,value in flow_counts.items():
    
    if randint(0,25)>23:
        print(str(key[0]) + ","+ str(key[1]) + "," + str(value)  )
#         print(str(key[0]) )
# #     print(key[1])
```
